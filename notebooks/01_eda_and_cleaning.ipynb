```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Oplossing 1: Gebruik het absolute pad (snel en werkt altijd)
df = pd.read_csv(r"C:\Users\DASEE\Downloads\creditcard_subset.csv")
```

```python
print(df.head)
```

```python
#Exploratory Data Analysis (EDA):
#We gaan nu de data verkennen en visueel inzicht krijgen:
# Stap 1: Klassenverdeling visualiseren


import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x="Class")
plt.title("Aantal frauduleuze en normale transacties")
plt.xticks([0, 1], ["Normaal", "Fraude"])
plt.show()
```

```python
 #Stap 2: Bekijk verdeling van 'Amount'
plt.figure(figsize=(10,4))
sns.histplot(data=df, x="Amount", bins=100, kde=True)
plt.title("Verdeling van transactiebedragen")
plt.xlabel("Bedrag")
plt.show()
```

```python
# Stap 3: Correlatieanalyse
corr = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Correlatie tussen alle variabelen")
plt.show()
```
